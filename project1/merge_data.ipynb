#### users dataframe and films dataframe have share the same column - 'BrukerID'. SO let join them on this column

```python
users_rankings = users.join(rankings.set_index('BrukerID'), on='BrukerID')
users_rankings
```

#### Now let join the above dataframe with films dataframe on 'FilmID'

```python
df = users_rankings.join(films.set_index('FilmID'), on='FilmID')
```
